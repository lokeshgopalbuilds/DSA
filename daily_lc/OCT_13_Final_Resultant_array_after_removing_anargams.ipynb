```python
from typing import List
```

```python
class Solution:
    def removeAnagrams(self, words: List[str]) -> List[str]:
        
        # Check Anaragams
        def is_anargam(src,desc):
            cnt = 0
            for i in zip(src,desc):
                if len(i[0]) == len(i[1]) == 1:
                    if i[0] == i[1]:
                        cnt+=1            
                elif src.count(i[0]) == desc.count(i[1]):
                    cnt+=1
                    
            
            if cnt == len(src) == len(desc):
                return True
            else:
                return False

        
        result = []
        for i in range(1, len(words)):
            src, desc = words[i],words[i-1]
            if not is_anargam(src,desc):
                result.append(words[i])
                
        return result
```

```python
words = ["abba","baba","bbaa","cd","cd"]
```

```python
from collections import Counter
def is_anagram(a: str, b: str) -> bool:
    return Counter(a) == Counter(b)
```

```python
# 1st word we are not going to compare so we will keep it as it as 

result = [words[0]]
for i in range(1,len(words)):
    
    src, desc = words[i], words[i-1]
    if not is_anagram(src,desc):
        result.append(src)
```
