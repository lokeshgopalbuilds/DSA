```python
def pythagorean_triplets(limit):
    triples = []
    for a in range(1, limit + 1):
        for b in range(a, limit + 1):
            c_sq = a * a + b * b
            c = int(c_sq ** 0.5)
            if c * c == c_sq and c <= limit:
                triples.append((a, b, c))
                triples.append((b, a, c))
    return triples

triplets = pythagorean_triplets(5)
for t in triplets:
    print(t)
```
