### Determinant
If it is 2X2 matrix use ad-bc <br>
If it is 3X3 matrix use a(ei-fh) - b(di-fg)+c(dh-eg)

```python
A = [[1, 2],
      [3, 4]]
```

```python
A = [[6, 1, 1],
      [4, -2, 5],
      [2, 8, 7]]
```

```python
def matrix_determinant(A):
    if len(A[0])==2:
        return A[0][0]*A[1][1]-A[0][1]*A[1][0]
    elif len(A[0])==3:
        det = (A[0][0]*(A[1][1]*A[2][2]-A[1][2]*A[2][1])
            -A[0][1]*(A[1][0]*A[2][2]-A[1][2]*A[2][0])
            +A[0][2]*(A[1][0]*A[2][1]-A[1][1]*A[2][0]))
        return det
matrix_determinant(A)
```

### Matrix Multiplication

```python
A = [[1, 2],
     [3, 4]]
B = [[5, 6],
     [7, 8]]
```

- First Matrix row representation 

```python
for i in range(len(A)):
    for j in range(len(A)):
        print(A[i][j], end=" ")
    print()
```

- Second Matrix col representation

```python
for i in range(len(B[0])):
    for j in range(len(B)):
        print(B[j][i], end=" ")
    print()
```

- Now Combined version

```python
result = []
for i in range(len(A)):
    row = []
    for j in range(len(B[0])):
        # Multiply row i of A with column j of B
        element = 0
        for k in range(len(B)):
            element += A[i][k] * B[k][j]
        row.append(element)
    result.append(row)
result
```

### Main Diagonal Sum 

```python
mat= [[8, -67, 48, -10, -23, -68, -28, 62, -94], [-50, 77, -39, -57, 15, 71, 44, 13, -53], [-30, 43, -92, -94, 42, 37, 95, 7, 19], [-1, 92, -85, 20, -53, -16, 31, -95, 4], [-62, 49, 0, 53, -93, -61, -49, 86, 97], [62, -26, -1, -38, 59, -93, -93, 12, -90], [78, 85, 90, 89, -65, 5, 48, 40, -86], [-79, -68, -89, 90, -40, 47, 68, -25, 92]]
```

```python
len(mat), len(mat[0])
```

```python
def diagonal_sum(mat):
    diagonal_sum = 0
    for i in range(len(mat[0])):
        for j in range(len(mat)):
            if i==j:
                # print(mat[i][j])
                diagonal_sum += mat[i][j]
            
    return diagonal_sum
```

```python
diagonal_sum(matrix)
```
