### Four Divisors

```python
# nums = [21, 4, 7]
# nums = [21,21]
# nums = [1,2,3,4,5]

nums = [7286,18704,70773,8224,91675]
```

- 1st Working Solution but TLE Error 

```python
# Traversing through each number
all_divisors = []
for i in nums:   
    each_number_divisors = []
    # checking how many divisors are there 
    for j in range(1,i+1):
        if i%j==0:
            each_number_divisors.append(j)
    
    if len(each_number_divisors) == 4:
        all_divisors.append(each_number_divisors)
    
ans = sum(sum(row) for row in all_divisors)
```

- Optimised One 

```python
all_divisors = []
for num in nums:
    divisor_count = 0
    divisor_sum = 0
    
    # Check divisors up to sqrt(num)
    i = 1
    while i * i <= num:
        if num % i == 0:
            divisor_count += 1
            divisor_sum += i
            
            # Add the paired divisor if it's different
            if i * i != num:
                divisor_count += 1
                divisor_sum += num // i
        i += 1
    
    if divisor_count == 4:
        all_divisors.append(divisor_sum)

ans = sum(all_divisors)
```
