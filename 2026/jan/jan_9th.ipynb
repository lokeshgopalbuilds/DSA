```python
arr1 = [2,3,1,3,2,4,6,7,9,2,19]
arr2 = [2,1,4,3,9,6]
```

```python
# arr1 = [28,6,22,8,44,17]
# arr2 = [22,28,8,6]
```

```python
left_over_elements = sorted([x for x in arr1 if x not in arr2])
arr1 = [i for i in arr1 if i not in left_over_elements]
result = sorted(arr1, key= arr2.index)
ans = result + left_over_elements
ans
```
