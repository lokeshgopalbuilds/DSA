#### Substring in given list of strings - Easy

```python
words = ["mass","as","hero","superhero"]

words = ["leetcode","et","code"]

words = ["blue","green","bu"]

words = ["leetcoder","leetcode","od","hamlet","am"]
```

```python
words.sort(key=len)
```

```python
# checkig each string is subset of other strings or not 
ans = []
for i in range(len(words)):
    
    # assignig sub string 
    sub_str = words[i]
    
    #print(sub_str)
    
    for j in words:
        if words[i]!=j:
            if words[i] in j:
                ans.append(sub_str)
                
list(set(ans))
```

### HTML Entity Parser - Medium

```python
from collections import OrderedDict

html_entities = OrderedDict([
    ("&quot;", '"'),
    ("&apos;", "'"),
    ("&amp;", "&"),
    ("&gt;", ">"),
    ("&lt;", "<"),
    ("&frasl;", "/")
])

html_entities
```

```python
html_entities.get('&amp')
```

```python
text = "&amp; is an HTML entity but &ambassador; is not."
```

```python
text = "and I quote: &quot;...&quot;"
```

```python
# Correct approach: replace entities in the text directly
result = text
for entity, char in html_entities.items():
    result = result.replace(entity, char)
    
print(result)
```
