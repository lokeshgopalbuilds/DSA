```python
nums = [1,3]
nums = [5,1,6]
nums = [3,4,5,6,7,8]
```

#### Below solution is accepting but worst Time complexity

- Reason : First we will calculate the combinations and then finding XOR is not good i guess

```python
from itertools import combinations

# all combinations list
all_combinations = []

# all combinations for each element
for r in range(len(nums) + 1):
    for combo in combinations(nums, r):
        all_combinations.append(list(combo))
        

#  XOR for each combination in all_combinations
combinations_sum = 0
for combo in all_combinations:
    if len(combo) == 0:
        combo_xor = 0
    else:
        combo_xor = combo[0]
        for i in range(1, len(combo)):
            combo_xor ^= combo[i]
    
    combinations_sum+=combo_xor

combinations_sum
```
